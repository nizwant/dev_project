```python
data_path = os.path.join("clustering-data-v1")
data_path = os.path.abspath(data_path)
data_path
```

# Test battery

```python
# Print the battery names
print(f"Battery names: {clustbench.get_battery_names(path=data_path)}")

# Print the dataset names for a specific battery
battery = "sipu"
print(f"Dataset names for battery '{battery}': {clustbench.get_dataset_names(battery, path=data_path)}")

# Load a specific dataset from a battery and print its details
dataset = "compound"
b = clustbench.load_dataset(battery, dataset, path=data_path)
# Print the dataset details
print(f"Dataset details: {b}")
print(f"Dataset shape: {b.data.shape}")
```

## Plot one of the datasets

```python
plt.figure(figsize=(12, 4))
for i in range(len(b.labels)):
    plt.subplot(1, len(b.labels), i + 1)
    genieclust.plots.plot_scatter(
        b.data,
        labels=b.labels[i]-1,
        axis='equal',
        title=f"Labels {i}",
    )
plt.show()
```
